Tabel `visits` (log/catatan server yang memuat data kunjungan ke situs web):

- `source_id` — ID sumber iklan
- `dt` — tanggal

```python
df_visits.groupby('session_date').agg({'uid': 'nunique'}).mean()
```

```python
dau_total = df_visits.groupby('session_date').agg({'uid': 'nunique'}).mean()
wau_total = df_visits.groupby('session_week').agg({'uid': 'nunique'}).mean()
mau_total = df_visits.groupby('session_month').agg({'uid': 'nunique'}).mean()

print('DAU for the entire period:', int(dau_total), 'users')
print('WAU for the entire period:', int(wau_total), 'users')
print('MAU for the entire period:', int(mau_total), 'users')
```

#### Seberapa sering pengguna kembali menggunakan produk?

```python
df_visits.head()
```

```python
first_visits = df_visits.groupby('uid').agg({'start_ts': 'min'}).reset_index()
first_visits.columns = ['uid', 'first_session']
first_visits.head()
```

```python
first_visits['first_session_dt'] = first_visits['first_session'].dt.date
first_visits['first_session_week'] = first_visits['first_session'].astype('datetime64[W]')
first_visits['first_session_month'] = first_visits['first_session'].astype('datetime64[M]')
first_visits['first_session_year'] = first_visits['first_session'].astype('datetime64[Y]')
```

```python
first_visits.head()
```

```python
visits_merged = pd.merge(df_visits, first_visits, on='uid')
visits_merged['age_days'] = (visits_merged['session_date'] - visits_merged['first_session_dt']).dt.days
visits_merged['age_month'] = round(visits_merged['age_days'] / 30,0).astype('int')
```

```python
visits_merged.head()
```

```python
visits_cohorts = visits_merged.pivot_table(index='first_session_month',
                                          columns = 'age_month',
                                          values = 'uid', 
                                          aggfunc = 'nunique')
```

```python
visits_cohorts.fillna('')
```

```python
# mengubah format cohort ke percentage
retention = pd.DataFrame()
for col in visits_cohorts.columns:
    retention = pd.concat([retention, visits_cohorts[col]/visits_cohorts[0]*100], axis=1)
retention.columns = visits_cohorts.columns
```

```python
retention.fillna('')
```

```python
retention[1].mean()
```

```python
retention[1].plot(kind='bar')
```

**findings:**
- rata-rata hanya 5.36% user yang kembali menggunakan produk pada bulan berikutnya

#### Berapa jumlah uang yang mereka kontribusikan? (LTV)

```python
first_buy.head()
```

```python
cohort_sizes = first_buy.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
```

```python
cohort_sizes
```

```python
cohorts_buy = pd.merge(df_orders, first_buy, how='inner', on='uid')
cohorts_buy_agg = cohorts_buy.groupby(['first_order_month', 'order_month']).agg({'revenue': 'sum'}).reset_index()
```

```python
cohorts_buy_agg.head()
```

```python
cohorts_buy_agg['age_month'] = (cohorts_buy_agg['order_month'] - cohorts_buy_agg['first_order_month']).dt.days
cohorts_buy_agg['age_month'] = round(cohorts_buy_agg['age_month'] / 30,0).astype('int')
```

```python
cohorts_buy_agg.head()
```

```python
cohorts_buy_merged = pd.merge(cohort_sizes, cohorts_buy_agg, on='first_order_month')
cohorts_buy_merged['ltv'] = cohorts_buy_merged['revenue'] / cohorts_buy_merged['n_buyers']
```

```python
cohorts_buy_merged.head()
```

```python
cohorts_ltv = cohorts_buy_merged.pivot_table(index='first_order_month',
                                          columns = 'age_month',
                                          values = 'ltv', 
                                          aggfunc = 'sum').cumsum(axis=1)
```

```python
cohorts_ltv.round(2).fillna('')
```

**findings:**
- berdasarkan `cohort_ltv`, cohort yang dimulai pada bulan `2017-09-01` merupakan cohort yang memiliki nilai kumulatif ltv tertinggi dengan nilai 13.44
- Rata-rata, setiap pelanggan dari cohort bulan `2017-09-01` menghasilkan pendapatan sebesar 13.44 selama waktu sembilan bulan mereka.

#### Seberapa menguntungkannya investasi tersebut? (ROI)

```python
month_revenue = df_orders.groupby(['uid','order_month']).agg({'revenue':'sum'}).reset_index()
```

```python
buyers = buyers.merge(month_revenue, on='uid')
```

```python
buyers.head()
```

```python
cohort_sizes_romi = buyers.groupby(['source_id', 'first_order_month']).agg({'uid':'nunique'}).reset_index()
cohort_sizes_romi.columns = ['source_id', 'first_order_month', 'n_buyers']
cohort_sizes_romi.head()
```

```python
cohort_revenue_romi = buyers.groupby(['source_id', 'first_order_month', 'order_month']).agg({'revenue':'sum'}).reset_index()
cohort_revenue_romi.columns = ['source_id', 'first_order_month', 'order_month', 'gp']
cohort_revenue_romi.head()
```

```python
report_romi = pd.merge(cohort_sizes_romi, cohort_revenue_romi, on=['source_id', 'first_order_month'])
report_romi.head()
```

```python
report_romi['age'] = (report_romi['order_month'] - report_romi['first_order_month']) / np.timedelta64(1, 'M')
report_romi['age'] = report_romi['age'].round().astype('int')
report_romi.head()
```

```python
report_romi['ltv'] = report_romi['gp'] / report_romi['n_buyers']
report_romi.head()
```

```python
costs_for_romi = costs.groupby(['source_id', 'costs_month']).agg({'costs':'sum'}).reset_index()
costs_for_romi
```

```python
report_romi = pd.merge(report_romi, costs_for_romi, left_on=['source_id', 'first_order_month'], right_on=['source_id', 'costs_month'])
report_romi.head()
```

```python
report_romi = report_romi.drop(['costs_month'], axis=1)
report_romi.head()
```

```python
report_romi['cac'] = report_romi['costs'] / report_romi['n_buyers'] 
report_romi['romi'] = report_romi['ltv'] / report_romi['cac']
report_romi.head()
```

```python
romi = report_romi.pivot_table(index='first_order_month',
                                          columns = 'age',
                                          values = 'romi', 
                                          aggfunc = 'mean').cumsum(axis=1)
```

```python
romi.round(2).fillna('')
```

**findings:**
- rata-rata cohort sudah paid off (mencapai angka 1) dan menghasilkan keuntungan, dengan cohort `2017-06-01` yang sudah mencapai angka 2 dengan waktu kurang lebih 1 tahun 
- beberapa cohort ada yang belum paid off, tapi jika diberi waktu beberapa bulan lagi, cohort-cohort tersebut kemungkinan akan paid off
